```python
import sympy as sym
```

```python
f1 = "8 + x"
f2 = "18 + 3*x"
z = "13 + 2*x"

x = sym.symbols("x")

p = sym.plotting.plot(f1, f2, z, (x, -20, 0), show=False)
p.ylim = (0, None)
p.show()
```

```python
f1 = "8 - x"
f2 = "18 - 3*x"
z = "13 - 2*x"

x = sym.symbols("x")

p = sym.plotting.plot(f1, f2, z, (x, 20, 0), show=False)
p.ylim = (0, None)
p.show()
```

```python
f1 = "-8 - x"
f2 = "-18 - 3*x"
z = "-13 - 2*x"

x = sym.symbols("x")

p = sym.plotting.plot(f1, f2, z, (x, -20, 0), show=False)
p.ylim = (None, 0)
p.show()
```
